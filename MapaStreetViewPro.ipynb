  *   **Lista com as Coordenadas obtidas pelas imagens:**

```python
def exif(local, imagem: str) -> dict:
    with open(f'/content/drive/MyDrive/Geo de Araras/Fotos Araras-{local}/DJI_{imagem}.JPG', 'rb') as image_file:
      my_image = Image(image_file)
      my_image.has_exif
      latitude = -1* (int(my_image.gps_latitude[0]) + ((my_image.gps_latitude[1]/60) + (my_image.gps_latitude[2]/3600)))
      longitude = -1* (int(my_image.gps_longitude[0]) + ((my_image.gps_longitude[1]/60) + (my_image.gps_longitude[2]/3600)))
      values = ('DJI_'+imagem,
                latitude,
                longitude, 
                my_image.gps_altitude)
      if local == "Local1":
          local1.append(values)
      elif local == "Local2":
          local2.append(values)
      else:
          print("No such file or directory")
```

# **🌎 Araras-SP:  *Local 1***


```python
map = folium.Map(location = 
                 [-22.321408, -47.378206], zoom_start = 17.45
                 )
folium.CircleMarker(location = [-22.321408, -47.378206],
                    radius = 150, 
                    color = '#87CEEB',
                    fill = True,
                    fill_color = '#87CEEB'
                    ).add_to(map)
map.add_child(folium.LatLngPopup())
ScrollZoomToggler().add_to(map)
map
```



# **Inserindo Marcadores StreetView: Local 1**



```python
maps1 = folium.Map(location = 
                   [-22.321408, -47.378206],
                   zoom_start = 20
                   )
```

```python
c = 0 
for lista in local1:
    tooltip = f"{local1[c][0]}"
    path = f"https://storage.googleapis.com/araras/local1/{local1[c][0]}.JPG"
    html = f'<i> <a href ={path} target="_blank"> Click to see picture: {local1[c][0]} </a> </i>'
    popup = folium.Popup(html, max_width=300)
    icon = folium.Icon(color="blue", icon="info-sign")
    marker = folium.Marker(location=[local1[c][1], local1[c][2]], popup=popup, tooltip=tooltip, icon=icon).add_to(maps1)
    c += 1
```

# **🌎 Araras-SP: *Visualizador StreetView Local 1***

```python
ScrollZoomToggler().add_to(maps1)
maps1
```
